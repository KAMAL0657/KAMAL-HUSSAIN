# Threshold Tuning & Class-Imbalance Numerics

Binary classification with imbalanced data (100 positives, 900 negatives).
We compute Precision, Recall, F1, and Accuracy for different thresholds.


```python

import pandas as pd

# Confusion matrix data
data = {
    "Threshold": [0.30, 0.50, 0.70],
    "TP": [82, 70, 55],
    "FP": [210, 90, 30],
    "FN": [18, 30, 45],
    "TN": [690, 810, 870]
}

df = pd.DataFrame(data)

# Metrics
df["Precision"] = df["TP"] / (df["TP"] + df["FP"])
df["Recall"] = df["TP"] / (df["TP"] + df["FN"])
df["F1"] = 2 * df["Precision"] * df["Recall"] / (df["Precision"] + df["Recall"])
df["Accuracy"] = (df["TP"] + df["TN"]) / (df["TP"] + df["FP"] + df["FN"] + df["TN"])

print("Metrics Table:")
display(df.round(3))

# Selection rule
best_val_f1 = df["F1"].max()
candidates = df[df["Recall"] >= 0.65]
selected = candidates.loc[candidates["F1"].idxmax()]

print("\nCandidates with Recall >= 0.65:")
display(candidates.round(3))

print("\nSelected Threshold:", selected["Threshold"])
```


## Conclusion
- Accuracy is misleading due to class imbalance
- F1-score is better for threshold tuning
- Threshold is chosen using Recall constraint + max F1
